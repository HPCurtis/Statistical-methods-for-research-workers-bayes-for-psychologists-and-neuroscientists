# Steps of Bayesian data analysis

<font size = "3"> Kruscke (2015) offers a step by step formulation for how to conduct a Bayesian analysis:

1. Identify the relevant data for the question under investigation.

2. Define the descriptive (mathematical) model for the data.

3. Specify the Priors for the model. If scientific research publication is the goal the priors will need to be accepted by a skeptical audience. This should be achievable using prior predictive checks to ascertain if the priors are reasonable.

4. Using Bayes rule estimate the posterior for the parameters of the model using the likelihood and priors. Then use the posterior for conducting your inferences.

5. Conduct model checks. i.e. Posterior predcitive checks.</font> 

<font size = "1">This notebook will follow this approach generally.</font> 

# Bayesian inference
<font size = "3"> Following on from that quick description of the classic between subjects t-test above its important to keep in mind that Bayesian inference is all derived from the application of Bayes rule $P(\theta \mid y) = \large \frac{P(y \mid \theta) \, P(\theta)}{P(y)}$ and as such while the following description of the Bayesian model is an equivalent to the two sample between t-test, It is fundamentally different, because it uses fully probabilistic modelling and the infernce is not based on sampling distributions.</font>
    
<font size = "1"> For a fuller description see the practicing Bayesian statistics markdown file within the Github repository.</font>

#  Step 1 - Identify the relevant data for the question under investigation

## Study/data description

The following notebook describes a Bayesian estimation analysis of the data from a replication study by Wagenmaker et al. (2015) of Topolisnki and Sparenberg (2012). The Wagenmaker et al. (2015)  are advocates of Bayesian analysis generally and in this replication advocate for the general application of Bayes Factor and for the use of their own analysis software JASP which interested readers can download for free https://jasp-stats.org/. Their replication demonstrated a Bayes factor analysis too demonstrate both the neccesity of replication and the power of Bayes to approach evidence of absence question that NHST cannot conduct.

The replication was based on Topolinski and Sparenbergs's second experiment. The original study and thus the replication study were premised on the conjecture that clockwise and counterclockwise motion are abstract symbols for progression/regression in time. Topolinski and Sparenberg predicted and aimed to show that clockwise movements would induce psychological states of temporal progression and oritation to the future.

In order to test the orignal finding and this conjecture Wagenmaker et al recruited 102 participants to take part in their replication of the original experiment. Participants were split into two conditions were they would either rotate a pair of kitchen rolls clockwise or counterclockwise. The shortened 12-item version of the openness to experience subscale of the Neuroticism–Extroversion–Openness Personality Inventory (NEO PI-R; Costa and McCrae, 1992) was completed by the participants.

However, following this description of the data this notebook is concerned with Bayesian estimation methods and demonstrates an example for analysing the repliciated data more in the vain of Kruschke (2013) Bayesian estiamtion supersedes the t-test.

# Step 2 - Define the descriptive statistical model \begin{align*}
y &\sim Normal(\mu_1,\sigma)
\\ y_{2} &\sim Normal(\mu_2,\sigma)
\\ \mu_{j} &\sim Normal(0,1) 
\\ \sigma &\sim HalfNormal(0,1)
\end{align*} 

```python
#Stan model put into string to be called when the model is compiled
student_t = """

data {
  int<lower=0> N1;  // # of observations (counter clockwise)
  int<lower=0> N2;  // # of observations (clockwise)
  // sample size is indetical in this case but of course but this is rare
  
  vector[N1] y1;  // Neo scores (counter clockwise)
  vector[N2] y2;  // Neo scores (clockwise);
  
  // Prior values
  int sigma_sigma;
  
}
transformed data{

  vector[N1] y1_std;
  vector[N2] y2_std;
  
  y1_std = (y1 - mean(y1)) / sd(y1);
  y2_std = (y2 - mean(y2)) / sd(y2);
} 

parameters {
  real mu_1;  // mean of (counter clockwise) group bounded to lower -2 and upper 2 in line with the likert
  //scale on the neo personality measure.
  real mu_2;  // mean of (clockwise) group 
  real<lower=0> sigma; // pooled standard deviation estiamted across conditions
}

model {

// Priors
mu_1 ~ std_normal();
mu_2 ~ std_normal();
sigma ~ normal(0, sigma_sigma);

// likelihoods
  y1_std ~ normal(mu_1, sigma);
  y2_std ~ normal(mu_2, sigma);
}

generated quantities {
  real diff;
  diff = mu_1 - mu_2;
}

"""
```

It important to note that the above model can be expressed more succintly, 
a vway to do this is in the One-way ANOVA notebook. The model formulation here is 
presented for pedagoligical reasons here.

```python
#Create python dictionary to specify the data to be modelled using th stan coded model above,
#(Should provalbly use length funtion for the N vector.
data = {'N1': len(df_counter) ,
               'N2': len(df_clock),             
               'y1': df_counter['mean_NEO'].values,
               'y2': df_clock['mean_NEO'].values,
               'sigma_sigma': 1}
```

```python
#Fit the model using 2000 sample iterations and 4 chains with the 4 cores on this machine.
#If you have more you can use more if less then use less.

sm = ps.build(student_t, data = data, random_seed =1)
```

# Fit Stan model

```python
Two_group_categorical_Regression = """
data{
int<lower=1> N;
int<lower=1> N1;
int<lower=1> N2;
int K;
// Input data in two speate vectors just to demonstrate vecotr concatentaion with Stan in simple case
vector[N1] y1;
vector[N2] y2;
matrix[N, K] X;

// Prior values 
int sigma_sigma;
}

transformed data{

vector[N1] y1_std;
vector[N2] y2_std;
vector[N] y_std;

// Standardise the two groups data
y1_std = (y1 - mean(y1)) / sd(y1);
y2_std = (y2 - mean(y2)) / sd(y2);

// Concatenate the datasets
y_std = append_row(y1_std, y2_std);

}

parameters{
vector[K] beta;
real<lower=0> sigma;
}

model{
// Priors 
// beta priors vectorised so same prior on intercept and slope
beta ~ std_normal();

// Standard deviation parameter
sigma ~ normal(0, sigma_sigma);

// Likelihood
// Estimation of all parameters through one likelihood function
y_std ~ normal(X * beta, sigma); 

}

"""
```

```python
# Order dataset for use in the dmatrix function below
ordered_df = df_clock.append(df_counter, ignore_index = True)
```

```python
#Using the patsy library allows for easy design matrix generation and then manipulation for use in Stan

# Create a treament coded design matrix
T_DM = np.asarray(dmatrix("Rotation ", ordered_df))
```

```python
# Create effect coded design matrix for two groups
Eff_DM = np.asarray(dmatrix("Rotation ", ordered_df))
Eff_DM[:, 1] = Eff_DM[:,1] - 0.5
```

```python
# Data place in python dictionary for treatment coded model
Two_group__Treat_categorical_Regression_dat = {'N': len(df_Mu_neo_condt),
                                               'N1': len(df_clock),
                                               'N2': len(df_counter),
                                               'K': np.shape(T_DM)[1],
                                               'y1': df_clock["mean_NEO"].values,
                                               'y2': df_counter["mean_NEO"].values,
                                               'X': T_DM, 
                                               'sigma_sigma': 1}
```

```python
# Data place in python dictionary for effect coded model
Two_group__Eff_categorical_Regression_dat = {'N': len(df_Mu_neo_condt),
                                               'N1': len(df_clock),
                                               'N2': len(df_counter),
                                               'K': np.shape(Eff_DM)[1],
                                               'y1': df_clock["mean_NEO"].values,
                                               'y2': df_counter["mean_NEO"].values,
                                               'X': Eff_DM, 
                                               'sigma_sigma': 1}
```

```python
Treat_m = ps.build(Two_group_categorical_Regression, data =  Two_group__Treat_categorical_Regression_dat, random_seed = 1)
```
