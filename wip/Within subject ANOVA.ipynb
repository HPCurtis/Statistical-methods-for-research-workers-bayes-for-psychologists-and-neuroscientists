# Steps of Bayesian data analysis

<font size = "3"> Kruscke (2015) offers a step by step formulation for how to conduct a Bayesian analysis:

1. Identify the relevant data for question under investigation.

2. Define the descriptive (mathematical) model for the data.

3. Specify the Priors for the model. In the case of scientific research publication is the goal, as such the priors must be accepted by a skeptical audience. Much of this can be achieved using prior predcitve checks to acsetain os the priors are reasonable.

4. Using Bayes rule estimate the posterior for the parameters of the model using the likelihood and priors. Then interprete and the posterior

5. Conduct model checks. i.e. Posterior predictive checks.</font> 

<font size = "1">This notebook will follow this approach generally.</font>

# Step 2 - define the descriptive model

As the The Bugs book (Lunn, Jackson, Best, Thomas, & Spiegelhalter, 2012) shows repeated measures ANOVA type models can be  estimated within the Bayesian framework through the use of multivariate normal likelihood. The multivariate normal model allows for the estimation of group means while also estimating and thus controlling for the correltaion between the repeated observed data values (the critical difference between between and within usubject ANOVA). The BUGS models use inverse wishsart priors for the covariance matrices when estimating correlation between data points. Stan, however, is not limited like Bugs in this way and the standard practice for stan users is to use LKJ priors (see, Lewandowski, Kurowicka, & Joe 2009, for further details). Additonally, the model below estimates the means using indicator variable formulation (McElreath, 2020) directly for comparisons of the differnces of the  $\mu's$ between the grouped observations across time instead of creating a regression slope as a regression line across time as done in the Bugs model (Lunn et al., 2012).

# \begin{align*}
y_i &\sim MVN(\mu_j,\sigma)
\\ \mu_{j} &\sim Normal(5, 1.5)
\\ \Omega &\sim LKJ(2)
\\ \sigma &\sim HalfNormal(0,1)
\end{align*} 

# Stan model

```python
# Cholsy factor model for effiecny and numerical atability
repeatedMeasuresANOVA = """ 
data{
int n;  
int k; 
// Data inputed in matrix form as the data is in wide format
vector[k] y[n]; 

int lkj_val;
real sigma_sigma;
real mu_mu;
real mu_sigma;
}

parameters{
vector<lower = 0>[k] mu;
vector<lower = 0>[k] sigma;

cholesky_factor_corr[k] omega;
}

model{

matrix[k,k] Sigma; 
Sigma = diag_pre_multiply(sigma, omega);

// priors
mu ~ normal(mu_mu, mu_sigma);
sigma ~ normal(0, sigma_sigma);
omega ~ lkj_corr_cholesky(lkj_val);

// likelihood
y ~ multi_normal_cholesky(mu,  Sigma);

}

generated quantities{
corr_matrix[k] rho = omega * omega';
}

"""
```

```python
y = dfReduced.iloc[:, 1:4].values 
data = {'n': len(dfReduced),
        'k': y.shape[1],
        'y': y,
        'lkj_val': 2,
        'sigma_sigma': .2,
        'mu_mu': 5.0,
        'mu_sigma':.5
       }
```

```python
sm = ps.build(repeatedMeasuresANOVA, data = data, random_seed = 1)
```

```python
%%capture
fit = sm.sample(num_chains= 4, num_samples = 1000, num_warmup = 1000, stepsize = 1)
```

The traceplots show some of the show the models has some divergences, the are only a few so this is not too problmatic but we know they are there and we must be considerate of the potetnial bias of posterios estimates. The result of these divergences could be many, in this example though it likely due to how correlated the data is (it is simulated by Sheffield University) and how few data points there are for the model to fit. However, if anyone reading this can see a much simpler explnation or better implementation please raise an issue explanation on the github repository.

Accounting for the issues above if we still wish to make inferences on these data we need to calculate contrasts between the posteriors of the group means. These contrasts of course are specified by the research questions these data were collected in a attempt to answer that being the change in cholesterol over time due to the consumption of margarine. The contrasts of interest then are Before - After4weeks, Before - After8weeks and After4weeks - After8weeks., which can be dome simple by taken the differnce between posterior sample of each groups $\mu$ parameter.

```python
# Calculate contrasts of interest
# Difference between Before and after 4 weeks
B_A4hdi = az.hdi(fit['mu'][0] - fit['mu'][1], hdi_prob = .95)

# Difference between Before and after 8 weeks
B_A8hdi = az.hdi(fit['mu'][0] - fit['mu'][2], hdi_prob = .95)

# Difference between after 4 weeks and after 8 weeks
A4_A8hdi = az.hdi(fit['mu'][1] - fit['mu'][2], hdi_prob = .95)
```

# Extra Bit

### Fitting an incorrect model to demonstrate why better model is better

Independent means Stan code.

```python
indpendent_model = '''
data{

int<lower = 0> N; // Number od data points
vector[N] y; // Dependent varaible data
int K; // number of groups
 // array of interger group indicator variables
int<lower = 1, upper = K> x[N];

}

parameters{

// By specifying the lower bound the normal for the priors
// become half normal prior distributions
vector[K] mu; 
vector<lower = 0>[K] sigma; 

}

model{

//priors 
mu ~ normal(5, 1.5); // priors on each group
sigma ~ normal(0,1);

//Likelihood
y ~ normal(mu[x], sigma[x]);
    
}
'''
```

```python
# Convert data to long format and generate group indicator variable for use in Stan (index starting at 1).

dfLong = pd.melt(dfReduced, id_vars=['ID'], value_vars=['Before', 'After4weeks', 'After8weeks'],
       var_name = 'Time', value_name = 'Cholesterol')
dfLong.Time = pd.Categorical(dfLong.Time, 
                             categories = ['Before', 'After4weeks', 'After8weeks'],
                             ordered = True)
dfLong['Time_cat'] = dfLong.Time.cat.codes
dfLong.Time_cat = dfLong.Time_cat + 1
```

```python
data_ind = {'y': dfLong['Cholesterol'].values,
            'N': len(dfLong['Cholesterol']),
            'x': dfLong['Time_cat'].values,
            'K': len(np.unique(dfLong['Time_cat']))}
```

```python
sm_ind = ps.build(indpendent_model, data = data_ind)
```

```python
fit_ind = sm_ind.sample(num_chains= 4, num_samples = 1000, num_warmup = 500, stepsize = .8 )
```

```python
az.summary(fit_ind)
```

```python
az.plot_forest(fit, var_names = 'mu', hdi_prob = .95, combined = True);

az.plot_forest(fit_ind, var_names = 'mu', hdi_prob = .95, combined = True);
```

```python
az.plot_trace(fit_ind, var_names='mu');
```

The fitting of the incorrect model shows the difference between the models in that for the multivariate normal model the $\mu$ posterior credible interval estimates are narrower than in the indenpendnet ANOVA style analysis. This due to accounting for the corrlation in the data. This can be viewed as the model making use of more information (covariance) in the data resulting in narrower (appropriate) estimates for the $\mu's$.

Final note - The indepedent model that ignores the correlation in the data does not result in diveregences when sampling, suggesting it is an easier model to sample generally and that the extreme correlations between the datapoints in this example is what makes the sampler struggle above.
