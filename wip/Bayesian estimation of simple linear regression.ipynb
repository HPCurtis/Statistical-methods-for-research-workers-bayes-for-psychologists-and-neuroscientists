#  Bayesian estimation equivalent of the simple linear regression

## The classic simple linear regression

Simple linear regression is the statistical method that is used for the estimation of linear associations between two continous variables. The statistical model is specified as $$y = \alpha + \beta x +\sigma$$

Under the classical Null hypothesis significance testing framework the null hypothesis is specified as $$H_0:\beta = 0 $$

and the alternative hypothesis is specified as $$H_A: \beta \neq 0$$

As with any NHST the alpha level should be choosen before the analysis or even better identified using a power analysis before data collection. Nevertheless, if $p < \alpha$ specified the NHST is declared as statistically signifinant and the null hypothesis can be rejected.

# Bayesian inference

<font size = "3"> Following the quick description of the classic simple linear regression above it is important to keep in mind that Bayesian inference is all derived from the application of Bayes rule $P(\theta \mid y) = \large \frac{P(y \mid \theta) \, P(\theta)}{P(y)}$ and as such while the following description of the Bayesian model is an equivalent to the classic simple regression, it is fundamentally different, because its uses fully probabilistic modelling and the infernces are not based on sampling distributions.</font>
    
<font size = "1"> For a fuller description see the Practicing Bayesian statistics markdown file within the Github repository.</font>

# Steps of Bayesian data analysis

<font size = "3"> Kruschke (2015) offers a step by step formulation for how to conduct a Bayesian analysis:

1. Identify the relevant data for the question under investigation.

2. Define the descriptive (mathematical) model for the data.

3. Specify the Priors for the model. If scientific research publication is the goal the priors must be accepted by a skeptical audience. Much of this can be achieved using prior predictive checks to ascertain if the priors are reasonable.

4. Using Bayes rule estimate the posterior for the parameters of the model using the likelihood and priors. Then use the posterior for conducting your inferences.

5. Conduct model checks. i.e. Posterior predcitive checks.</font> 

<font size = "1">This notebook will follow this approach generally.</font> 

#  Step 1 - Identify the relevant data for question under investigation

## STUDY DESCRIPTION

Reasoning about cause and effect is fundamental to being human. However this reasoning can often take the form of thinking using heurtistics, as analysing all potetential casues is cumbersome cognitively or outright impossible. These heuristics may generate biases, which they can then potentially use to explain the "staus quo" for the way things are. People who reason from these heuristics then are at risk of conflating the way something "is" to the way it "ought" to be.

Investigating this was the focus of the study by Tworek and Cimpian (2016). They approach their investigation by quantifiying and recording the inherent bias of a sample of individuals. They then evaluated is these biases were related to how they explained if the outcomes of events were desriable for themselves.

122 participants were acquired. Each participant read 6 staus quo articles for societal practices. The participant then answered questions that assesed the level that the societal practices ought to happen: this measured the level of ought inferences. Participants then completed a 15 item questionaire designed to measure inherent bias. Finally, participants also completed a series of other tests for intelligence, political beliefs, belief in a just world. In additon to describing a range of general demographics (e.g. level of education).

# Step 2 - Define the descriptive statistical model \begin{align*}
y_i &\sim Normal(\mu_i, \sigma) 
\\ \mu_i &= \beta_0 + \beta_1(x-\bar{x})
\\ \beta_0 &\sim normal(5,1)
\\ \beta_1 &\sim normal(0,.5)
\\ \sigma &\sim halfcauchy(0, .1) 
\end{align*} 

<font size = "3">The formulation for presenting statistical models here follows that used by McElreath (2020) for its intuitive nature. In plain english the model specifies that the dependent variable $y_i$ is distributed normally in terms of the Likelihood with the $\mu_i = \beta_0 + \beta_1(x-\bar{x})$ and $\sigma$ to be estimated. With the prior for the $\beta_0$ parameter being $Normal(0.5, 0.2)$ followed by the $\beta_1$ parameter being $Normal(0.5, 0.2)$ and the $\sigma$ being $halfcauchy(0, .1)$</font>

# Step 3 - Specifying priors

## Prior predictive checks

<font size = "3">The model above may seem arbitary, but it is informed by the prior predictive checks below which determined that the priors specifed are reasonable for model parameter values for the outcome space of data that Tworek and Cimbian (2016) collected, when analysed by simple lnear regression model.

The neccessity for these prior predictive checks becomes essential with increasing model complexity. As such models create surprising interactions between parameters and thus lead to surprising data the model is able to generate and the explicit assumptions that your model makes about the data generating process prior to observing any data.
</font>

### Visualising priors

```python
#Simulate 100 intercept and slope parameters based on the model priors
np.random.seed(1)
N = 100
beta_0 = np.random.normal(loc = 5, scale = 1, size = N)
beta_1= np.random.normal(loc = 0, scale = .5,size = N)
```

```python
# Plot the lines based on the priors
x = np.linspace(0,10,100)
xbar = np.mean(filtered['Inherence_Bias'])
for i in range(N):
    y = beta_0[i] + beta_1[i] * (x - xbar)
    plt.plot(x, y, '-y', alpha =.1)
plt.xlim(0,10)
plt.ylim(-5,15)

# Plot horizontal lines for maximum  and minimum scores for the dependent variable
plt.hlines(0, xmin = 0 , xmax = 10, linestyles= 'dashed', colors= 'r');
plt.hlines(10, xmin = 0 , xmax = 10, linestyles= 'dashed', colors= 'r');
```

Now simulate data from the simulated parameter from the statistical model defined above.

```python
# Set seed for reproducibilty of result
np.random.seed(1)
sigma = stats.halfcauchy.rvs(size = N, loc=0, scale= .1)
sim_data = np.random.normal(loc = beta_0 + beta_1 * (x - xbar), scale = sigma)

sns.displot(sim_data);
```

All together the prior predictive checks show that these Weakly informative priors that result in skeptical parameter for $\beta_0$ estimate and that a postive and negative relationship with the $\beta_1$ is credilble before seeing the data.

# Step 4 - Use Bayes rule

## Stan model of Bayesian simple regression

```python
Simple_Regression_model = """ 

data {

int<lower=0> N; // Number of data points
int K; // Number of predictors
matrix[N, K] x; // Specfiying design matrix for the regression intercept and predictor values
vector[N] y; // vector of the dependent variable data points
  
}

transformed data{
// Creating a new design matrix with the Independent variable of Inherence bias
// centered. This may potentially be simpler to do outside of Stan but it demonstrates
// some of the options of Stan
matrix[N, K] x_transformed = x;
x_transformed[,2] = x[,2] - mean(x[,2]);

}

parameters {

// Specify simple regression model parameters specified.
vector[K] beta; // vector of beta coefficent that includes the intercept term
real<lower=0> sigma; // Standard deviation
  
}


model {

// Specify priors for the model parameters.
// priors
beta[1] ~ normal(5, 2);
beta[2] ~ normal(0,.5);
sigma ~ cauchy(0, .1); 

//likelihood
y ~ normal(x_transformed * beta, sigma);
  
}

generated quantities {

// Simulate data sets from the mdoel for posterior predicitve checks
real yrep[N];
yrep = normal_rng(x_transformed * beta, sigma);

real effect_size;
effect_size = beta[1]/sigma;
}

"""
```

```python
# Generate the design matrix for simple regression which is then convert to numpy array 
# to pass to the data block of Stan.
x = pt.dmatrix(" ~ Inherence_Bias", data = filtered)
x = np.asarray(x)

# Create python dictionary to pass to Stan data block
data = {'N': len(filtered),
        'y': filtered["Ought_Score"].values,
        'x': x,
        'K': x.shape[1]
       }
```

```python
# Compile stan model to C++ code to sample posterior for the model below
sm  = ps.build(Simple_Regression_model, data = data, random_seed = 1)
```

```python
# Fit and sample the model parameters.
fit = sm.sample(num_chains = 4, num_samples = 2000)
```

```python
fit_df = fit.to_frame()
az.summary(fit, var_names = ['beta','sigma', 'effect_size'])
```

# Post model fit-visualisations - Bayesian simple regression

## Posterior distribution plots

```python
# Posterior plots showing the default 94% Credible interval of the arviz package
az.plot_posterior(fit, var_names = ["beta", "sigma"]);
```

## Posterior  autocorrelation plots

```python
# Using the arviz package the autocorrelation of the 4 MCMC chains can be plotted.
az.plot_autocorr(fit,  var_names=("beta", "sigma"));
```

The autocorrelation plots do not show any serious autocorrelation problems, as the values quickly decrease to 0.

# Posterior trace plots 

```python
# Trace plots
az.plot_trace(fit, var_names = ["beta", "sigma"]);
```

The traceplot show good mixing of chains and show "Fuzzy catepillars".

# Step 5 - Posterior predictive checks

# Maximum aposterior estimate (MAP) for the fitted regression model 

```python
x_vals = np.linspace(1,10,100)
y_vals = np.mean(fit_df['beta.1']) + np.mean(fit_df['beta.2']) * (x_vals - xbar)
sns.scatterplot(data = filtered, x = 'Inherence_Bias', y = 'Ought_Score', alpha = .9);
plt.plot(x_vals, y_vals, '-');
```

The MAP is the most probable result, but it ignores the uncertainty in this estimate, so below the other credible regression lines within the posterior are plotted

# Uncertainty in the regression estimates

```python
for i in range(100, 2000):
    # Mean centred
    y_vals = fit_df['beta.1'][i] + fit_df['beta.2'][i] * (x_vals - xbar)
    plt.plot(x_vals, y_vals, '-y', alpha=0.01)

sns.scatterplot(data = filtered, x = 'Inherence_Bias', y = 'Ought_Score', alpha = .9);
```

# Posterior predictive checks

```python
idata = az.from_pystan(posterior=fit, posterior_model = sm, posterior_predictive = ['yrep'], observed_data= 'y')

# Plot posterior simulated data sets for posterior predictive check
az.plot_ppc(idata , data_pairs = {"y": 'yrep'}, num_pp_samples = 100);
```

The Posterior predictive checks above suggest that the model was able to capture the observed data fairly well and suggests no serious issues with the model used to analyse the data.  

# Reporting the results of the Bayesian simple regression

In addition, as Gabry et al. (2019) argue visualisations maybe even more key so the all the visualisations above would have to be included with any write up. Anyway the write up below generally follows the advice of Krushcke (2015) chapter 25. In any application though it comes down to the problem to be described an the audience that needs to be convinced. </p><br/>

<h2>Write up</h2><br/>

Four MCMC chains were ran each taking 2000 samples from the posterior. 1000 of the samples from each chain were used as warmup and discarded from the final analysis (i.e not used in the inferences). Convergence of the MCMC chains was examined through the use of autocorrelation and traceplots. Visual inspection suggested no issues in terms of autocorrelation or chain mixing and the generally accepted metric of r-hat < 1.05 and an effective sample size > 400 was achieved for all estimated parameters (see, Vehtari et al. for further discussion).

The parameter of interest in this data analysis was the $\beta_1$ parameter. The posterior for $\beta_1$ showed that there was positive association between inherent bias scores and Ought scores $\mu$ = 5.67, with a 94% CrI [0.132,	0.443], as such this result is consisent with claim of statistical significance under the Bayesian framework (Kruschke, 2018; Makowski, Ben-Shachar, Chen, & Lüdecke, 2019).
